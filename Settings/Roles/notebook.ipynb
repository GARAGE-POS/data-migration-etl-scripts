```python
import pandas as pd
```

```python
df = pd.read_csv('Roles.csv')
df
```

```python
df['GroupName'] = df['GroupName'].map(lambda x: x.strip().lower())
```

### Valid Users

```python
groups = df.groupby('UserID').groups

valid_users = []

for k,v in groups.items():

    group_names = set(df.loc[v,'GroupName'].values)
    print(f"{k}: {group_names}")
    # is_user_valid = group_names.issubset({'manager', 'cashier', 'technician', 'مدير', 'كاشير', 'فني', 'المدير'})
    is_user_valid = group_names.issubset({'manager', 'cashier', 'technician', 'مدير', 'كاشير', 'فني', 'المدير', 'الفني', 'الكاشير'})
    print(is_user_valid)
    print()
    if is_user_valid:
        valid_users.append(k)
```
